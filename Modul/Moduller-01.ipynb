## Modül Kavramı 

### Modül Nedir? Ne İşimize Yarar

Pythonda yazdığımız her bir farklı dosya bir modül kabul edilir. Modüller içerisinde yazdığımız kodu, fonksiyonları, sınıfları daha sonra başka programlarda veya aynı programın farklı yerlerinde çağırabilir ve kullanabiliriz.

### Modül Ekleme

Daha önce yazdığımız bir modülü herhangi bir kod dosyasında tekrar kullanabilmek için, modülü kod dosyasına import keywordü ile dahil etmeliyiz.

```python
import math # Math Modülünü bu py dosyasına dahil ettim
import time # Time Modülünü bu py dosyasına dahil ettim
import os   # OS Modülünü bu py dosyasına dahil ettim
import sys # Sys Modülünü bu py dosyasına dahil ettim
```

### Modülün Belirli Bir Kısmını Ekleme 

Syntax Yapısı

from modul_ismi import fonksiyon_ismi 

```python
from os import name
```

```python
name
```

### Modülün İçerisindeki Tüm Fonksiyonları Ekleme

Syntax Yapısı

from modul_ismi import *

```python
from os import *
```

### Modül Eklerken Modülllere İsim Verme

Modülü kod bloğuna dahil ederken as keywordü ile kendi belirlediğimiz ismi verebiliriz.

```python
import math as matematik
import os as isletimsistemi
import sys as sistem
```

### Modül İçerisindeki Fonksiyonları Görme

dir() fonksiyonunu kullanarak modül içerisindeki fonksiyonları listeyebiliriz.

```python
dir(matematik)
```

```python
dir(os)
```

### Modül İçerisindeki Fonksiyonların Ne İşe Yaradığını Görme

help() fonksiyonunu kullanarak modül içerisindeki fonksiyonların işlevlerini listeyebiliriz.

```python
help(math)
```
