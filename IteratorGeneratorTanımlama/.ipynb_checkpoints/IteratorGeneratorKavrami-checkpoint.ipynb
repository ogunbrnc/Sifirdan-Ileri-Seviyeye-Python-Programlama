#### Kendi Sınıfımız İçin __next__() ve __iter__() Fonksiyonları 

```python
class Sayi:
  def __iter__(self):
    self.num = 0
    return self

  def __next__(self):
    temp = self.num
    self.num += 1
    return temp

sayi = Sayi()
sayiIt = iter(sayi)

print(next(sayiIt))
print(next(sayiIt))
print(next(sayiIt))
print(next(sayiIt))
```

#### Kendi Sınıfımız İçin StopIteration Hatası Tanımlama

```python
class Sayi:
  def __iter__(self):
    self.num = 0
    return self

  def __next__(self):
    if self.num<5:
        temp = self.num
        self.num += 1
        return temp
    else:
        raise StopIteration
        
```

```python
sayi = Sayi()
sayiIt = iter(sayi)

while True:
    print(next(sayiIt))
```

```python
sayi = Sayi()
sayiIt = iter(sayi)

while True:
    try:
        print(next(sayiIt))
    except:
        break
```

### Generator Nedir? Neden İhtiyacımız Var?

Normal fonksiyonlardan farklı olarak, geriye bir iterator dönen fonksiyonlardır. Generator fonksiyonlar, return ifadenin yerine kullanılacak en az 1 tane yield ifade içermelidir. 

__iter__ , __next__() ve StopIteration otomatik olarak oluşturulur.

```python
def deger_arttir():
    sayi = 0
    print("Ilk Deger:",sep=" ")
    yield sayi

    sayi += 1
    print("Ikinci Deger:",sep="")
    yield sayi

    sayi += 1
    print("Ucuncu Deger:",sep="")
    yield sayi
```

```python
deger_arttir_iter=deger_arttir()
```

```python
print(next(deger_arttir_iter))
print(next(deger_arttir_iter))
print(next(deger_arttir_iter))
```

```python
print(next(deger_arttir_iter))
```

```python
while True:
    try:
        print(next(deger_arttir_iter))
    except:
        break
```

#### String İfade Döndüren Generator

```python
def kup_yazdir():
    sayiListesi=[i ** 3 for i in range(50)]
    for j in range(50):
        yield "{}^3:{}".format(j,sayiListesi[j])
    
```

```python
for i in kup_yazdir():
    print(i)
```

#### () ile Generator Tanımlama

```python
kare_liste = [x**2 for x in range(20)]
print(kare_liste)
```

```python
generator = (x**2 for x in range(20))
print(generator)
```

```python
for i in generator:
    print(i)
```
