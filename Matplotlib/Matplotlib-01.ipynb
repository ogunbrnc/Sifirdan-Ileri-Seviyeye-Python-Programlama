## Matplotlib ile Veri Görselleştirme

### Neden İhtiyacımız Var?
Matplotlib kütüphanesi içerisinde bulunan pyplot sayesinde verileri grafiklere dökebiliyor ve daha okunulabilir,anlaşılır hale getirebiliyoruz.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.arange(1,11)
y=np.random.randint(1,50,size=(10,1))
y
```

### Grafik Çizimi

```python
plt.plot(x,y,"green")
plt.show()
```

```python
plt.plot(x,y,"green")
plt.xlabel("X Değerleri")
plt.ylabel("Y Değerleri")
plt.title("X-Y Grafiği")
plt.show()
```

### Grafiği Alt Parçalara Bölmek (Subplot)

plt.subplot(satir_sayisi,sutun_sayisi,index)

```python
plt.subplot(1,2,1)
plt.plot(x,y,"red")
plt.subplot(1,2,2)
plt.plot(x,y,"blue")
```

```python
plt.subplot(2,2,1)
plt.plot(x,x**2,"red")
plt.subplot(2,2,2)
plt.plot(x,x**3,"blue")
plt.subplot(2,2,3)
plt.plot(x,x**4,"black")
plt.subplot(2,2,4)
plt.plot(x,x**5,"pink")
```

### Figure ve Axes Oluşturma

add_axes([left, bottom, width, height])

```python
fig=plt.figure(figsize=(2,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,"red")
axes.set_xlabel("X Değerleri")
axes.set_ylabel("Y Değerleri")
axes.set_title("X vs Y Grafiği")

axes2=fig.add_axes([0,1,2,1])
axes2.plot(x,y,"yellow")
axes2.set_xlabel("X Değerleri")
axes2.set_ylabel("Y Değerleri")
axes2.set_title("X vs Y Grafiği")

plt.show()
```

### Figure ve SubPlots

plt.subplots(nrows,ncols,figsize)

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(4,4))
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(4,4))
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(4,4))

axes[0].plot(x,y,color="red")
axes[0].set_title("Axes 0")

axes[1].plot(x,y,color="green")
axes[1].set_title("Axes 1")


plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(4,4))

axes[0][0].plot(x,y,color="red")
axes[0][0].set_title("Axes 0")

axes[0][1].plot(x,y,color="green")
axes[0][1].set_title("Axes 1")

axes[1][0].plot(x,y,color="yellow")
axes[1][0].set_title("Axes 2")

axes[1][1].plot(x,y,color="pink")
axes[1][1].set_title("Axes 3")

plt.tight_layout()
```

### Axes Stil Kazandırma

linestyle: çizginin şeklini belirtir.

linewidth: çizginin kalınlığını belirtir.

marker: değerlerin eşleştiği noktanın şeklini belirtir.

markeredgecolor: değerlerin eşleştiği noktanın rengini belirtir.

markeredgewidth: değerlerin eşleştiği noktanın kalınlığını belirtir.

```python
fig,axes=plt.subplots(figsize=(6,6))

axes.plot(x,x,label="X",color="red",linewidth=2,linestyle="-",marker="o",markeredgecolor="black",markeredgewidth=5)
axes.plot(x,x**2,label="X^2",color="blue",linewidth=3,linestyle="--",marker="v",markeredgecolor="black",markeredgewidth=5)
axes.plot(x,x**3,label="X^3",color="yellow",linewidth=4,linestyle="-.",marker=".",markeredgecolor="black",markeredgewidth=5)
axes.plot(x,x**4,label="X^4",color="orange",linewidth=5,linestyle=":",marker="^",markeredgecolor="black",markeredgewidth=5)


axes.legend()
```

### Pasta Grafik

```python
harfler=["A","B","C","D","E","F"]
sayilar=[25,20,35,25,15,45]

fig=plt.figure(figsize=(5,5))
plt.pie(sayilar,labels=harfler)
plt.show()
```

### Bar Grafik

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,1,1])

harfler=["A","B","C","D","E","F"]
sayilar=[25,20,35,25,15,45]

ax.bar(harfler,sayilar)
plt.show()

fig.savefig("ornekfigure.png")
```

### Area Plot

```python
notlar={"Ogrenci1":[65,85,75,95],"Ogrenci2":[45,55,80,90]}
dersler=["Turkce","Matematik","Fizik","Kimya"]
dataFrame=pd.DataFrame(notlar,index=dersler)
dataFrame.plot(kind="area",stacked=False)
plt.show()
```
